```python
# reso.set_index('resort_name')
# reso.plot(reso.loc[:0, ['green_percent', 'blue_percent', 'black_percent']])
# vail = reso[reso['resort_name'] == 'Vail, CO']
# plt.plot(vail['green_percent'], vail['blue_percent'])
ham = []
def finder(x):
    ham.append(reso.loc[reso['resort_name']==x, 'green_percent'].iloc[0] *100)
    ham.append(reso.loc[reso['resort_name']==x, 'blue_percent'].iloc[0] *100)
    ham.append(reso.loc[reso['resort_name']==x, 'black_percent'].iloc[0] *100)
    

```

```python
finder('Vail, CO')
print(ham)
```

```python
plt.pie(ham, labels=('Green Runs', 'Blue Runs', 'Black Runs'), colors=('green', 'blue', 'black'), autopct="%1.2f%%")
```
